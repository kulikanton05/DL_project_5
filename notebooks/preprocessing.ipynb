```python
df['Risk'] = np.where((df['Company response to consumer'] == 'Closed with monetary relief') |
    (df['Company response to consumer'] == 'Closed with non-monetary relief') |
    (df['Timely response?'] == 'No'), 1, 0)
df
```

```python
df['Risk'].value_counts()
```

```python
df['Tags'].value_counts()
```

```python
df["Complaint_length"] = df["Consumer complaint narrative"].str.len() # Считаем длину обращения
df["Word_count"] = df["Consumer complaint narrative"].str.split().str.len() # Считаем количество слов в обращении
df
```

```python
df["Date received"] = pd.to_datetime(df["Date received"])
df
```

```python
df["month"] = df["Date received"].dt.month
df["day_of_week"] = df["Date received"].dt.day_name()
df["hour"] = df["Date received"].dt.hour
df
```

```python
cols_to_drop = [
    "Date sent to company",
    "Company response to consumer",
    "Timely response?",
    "Complaint ID",
    "ZIP code"
]
df = df.drop(columns=cols_to_drop)
df
```

```python
cols_to_drop1 = [
    "Submitted via",
    "Tags"
]
df = df.drop(columns=["Submitted via", "Tags"])
df
```

```python
df.isnull().sum()
```

```python
df = df.drop('Company public response', axis=1)
df
```

```python
df = df.drop(columns=['Sub-product', 'Sub-issue'])
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df[df.duplicated(keep=False)]
```
